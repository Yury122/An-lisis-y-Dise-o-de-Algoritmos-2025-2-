```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data  # variables (longitud y ancho de sépalos y pétalos)
y = iris.target  # etiquetas de clase (setosa, versicolor, virginica)
target_names = iris.target_names
```

```python
X_std = StandardScaler().fit_transform(X)
# X_std
```

```python
# Mantener 2 componentes principales
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)

print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)
print("Varianza total explicada (2 componentes):", sum(pca.explained_variance_ratio_))
```

```python
colors = ['slateblue', 'tomato', 'springgreen']

plt.figure(figsize=(8,6))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, alpha=.7, lw=2,
                label=target_name)

plt.legend()
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.title("PCA del dataset Iris")
plt.show()
```

# **El análisis PCA aplicado al conjunto de datos Iris permitió reducir las cuatro variables originales a dos componentes principales que explican más del 95% de la variabilidad total. El primer componente, fuertemente influenciado por las medidas de pétalos, separa con claridad a la especie Iris setosa, mientras que el segundo, relacionado con las medidas de sépalos, aporta información adicional para distinguir entre Iris versicolor y Iris virginica, aunque con cierta superposición. En conjunto, el PCA muestra que los pétalos son las variables más determinantes en la diferenciación de especies y que esta técnica resulta una herramienta poderosa para simplificar la dimensionalidad, facilitar la visualización y destacar los patrones principales en datos biológicos.**
