```python
import numpy as np
import math

# --- Paso 1: Definir frecuencias manualmente del dataset de tenis ---

# Conjunto total
total = 14
play_yes = 9
play_no = 5

# --- Paso 2: Entropía total de "Play" ---

p_yes = play_yes / total
p_no = play_no / total

H_S = -(p_yes * math.log2(p_yes) + p_no * math.log2(p_no))

print(f"Entropía total H(S): {H_S:.3f}")  # Resultado esperado: 0.940

# --- Paso 3: Atributo: OUTLOOK ---
# Sunny: 5 ejemplos (2 Yes, 3 No)
# Overcast: 4 ejemplos (4 Yes, 0 No)
# Rain: 5 ejemplos (3 Yes, 2 No)

# Sunny
sunny_total = 5
sunny_yes = 2
sunny_no = 3
p1 = sunny_yes / sunny_total
p2 = sunny_no / sunny_total
H_sunny = -(p1 * math.log2(p1) + p2 * math.log2(p2))

# Overcast
H_overcast = 0  # Solo "Yes"

# Rain
rain_total = 5
rain_yes = 3
rain_no = 2
p1 = rain_yes / rain_total
p2 = rain_no / rain_total
H_rain = -(p1 * math.log2(p1) + p2 * math.log2(p2))

# Entropía condicional de Outlook
H_outlook = (sunny_total/total) * H_sunny + \
            (4/total) * H_overcast + \
            (rain_total/total) * H_rain

IG_outlook = H_S - H_outlook

print(f"\nAtributo: Outlook")
print(f"  H(Sunny) = {H_sunny:.3f}")
print(f"  H(Overcast) = {H_overcast:.3f}")
print(f"  H(Rain) = {H_rain:.3f}")
print(f"  H(Play | Outlook) = {H_outlook:.3f}")
print(f"  IG(Outlook) = {IG_outlook:.3f}")  # Esperado: ~0.246

# --- Paso 4: Atributo: TEMPERATURE ---
# Hot: 4 (2 Yes, 2 No)
# Mild: 6 (4 Yes, 2 No)
# Cool: 4 (3 Yes, 1 No)

# Hot
hot_total = 4
hot_yes = 2
hot_no = 2
p1 = hot_yes / hot_total
p2 = hot_no / hot_total
H_hot = -(p1 * math.log2(p1) + p2 * math.log2(p2))

# Mild
mild_total = 6
mild_yes = 4
mild_no = 2
p1 = mild_yes / mild_total
p2 = mild_no / mild_total
H_mild = -(p1 * math.log2(p1) + p2 * math.log2(p2))

# Cool
cool_total = 4
cool_yes = 3
cool_no = 1
p1 = cool_yes / cool_total
p2 = cool_no / cool_total
H_cool = -(p1 * math.log2(p1) + p2 * math.log2(p2))

# Entropía condicional
H_temp = (hot_total/total)*H_hot + (mild_total/total)*H_mild + (cool_total/total)*H_cool
IG_temp = H_S - H_temp

print(f"\nAtributo: Temperature")
print(f"  H(Hot) = {H_hot:.3f}")
print(f"  H(Mild) = {H_mild:.3f}")
print(f"  H(Cool) = {H_cool:.3f}")
print(f"  H(Play | Temp) = {H_temp:.3f}")
print(f"  IG(Temp) = {IG_temp:.3f}")  # Esperado: ~0.028

# --- Paso 5: Atributo: HUMIDITY ---
# High: 7 (3 Yes, 4 No)
# Normal: 7 (6 Yes, 1 No)

# High
high_yes = 3
high_no = 4
p1 = high_yes / 7
p2 = high_no / 7
H_high = -(p1 * math.log2(p1) + p2 * math.log2(p2))

# Normal
normal_yes = 6
normal_no = 1
p1 = normal_yes / 7
p2 = normal_no / 7
H_normal = -(p1 * math.log2(p1) + p2 * math.log2(p2))

H_humidity = (7/14)*H_high + (7/14)*H_normal
IG_humidity = H_S - H_humidity

print(f"\nAtributo: Humidity")
print(f"  H(High) = {H_high:.3f}")
print(f"  H(Normal) = {H_normal:.3f}")
print(f"  H(Play | Humidity) = {H_humidity:.3f}")
print(f"  IG(Humidity) = {IG_humidity:.3f}")  # Esperado: ~0.152

# --- Paso 6: Atributo: WIND ---
# Weak: 8 (6 Yes, 2 No)
# Strong: 6 (3 Yes, 3 No)

# Weak
weak_yes = 6
weak_no = 2
p1 = weak_yes / 8
p2 = weak_no / 8
H_weak = -(p1 * math.log2(p1) + p2 * math.log2(p2))

# Strong
strong_yes = 3
strong_no = 3
p1 = strong_yes / 6
p2 = strong_no / 6
H_strong = -(p1 * math.log2(p1) + p2 * math.log2(p2))

H_wind = (8/14)*H_weak + (6/14)*H_strong
IG_wind = H_S - H_wind

print(f"\nAtributo: Wind")
print(f"  H(Weak) = {H_weak:.3f}")
print(f"  H(Strong) = {H_strong:.3f}")
print(f"  H(Play | Wind) = {H_wind:.3f}")
print(f"  IG(Wind) = {IG_wind:.3f}")  # Esperado: ~0.048
```

```python
# --- Paso 1: Subir el archivo CSV ---
from google.colab import files
uploaded = files.upload()

# --- Paso 2: Cargar el dataset ---
import pandas as pd
import numpy as np

# Cargar el archivo (ajusta el nombre si es diferente)
df = pd.read_csv("tennis_dataset.csv")
print("Dataset cargado:\n")
print(df)

# --- Función para calcular entropía ---
def entropy(column):
    """
    Calcula la entropía de una columna categórica.
    """
    values, counts = np.unique(column, return_counts=True)
    probabilities = counts / counts.sum()
    return -np.sum(probabilities * np.log2(probabilities))

# --- Función para calcular ganancia de información ---
def information_gain(data, feature, target="Play"):
    """
    Calcula la ganancia de información de un atributo respecto al objetivo.
    """
    HS = entropy(data[target])  # entropía del conjunto
    values, counts = np.unique(data[feature], return_counts=True)

    # Entropía condicional
    weighted_entropy = np.sum([
        (counts[i]/np.sum(counts)) * entropy(data[data[feature]==values[i]][target])
        for i in range(len(values))
    ])

    return HS - weighted_entropy

# Función para calcular entropías condicionales ---
def conditional_entropies(data, feature, target="Play"):
    """
    Imprime la entropía condicional del target para cada valor del feature dado.
    """
    print(f"\nEntropías condicionales de '{target}' dado '{feature}':")
    values = data[feature].unique()
    for v in values:
        subset = data[data[feature] == v][target]
        e = entropy(subset)
        print(f"  Entropía({target} | {feature} = {v}): {e:.3f}")

# --- Paso 3: Calcular entropía del conjunto ---
HS = entropy(df["Play"])
print(f"\nEntropía total del conjunto objetivo 'Play': {HS:.3f}\n")

# --- Paso 4: Calcular ganancia y entropías condicionales para cada atributo ---
features = ["Outlook", "Temperature", "Humidity", "Wind"]
for f in features:
    print(f"Atributo: {f}")

    # Mostrar entropías condicionales por valor del atributo
    conditional_entropies(df, f)

    # Mostrar ganancia de información
    ig = information_gain(df, f, target="Play")
    print(f"  ➤ Ganancia de información para '{f}': {ig:.3f}\n")
```

```python
import sympy as sp

# Número de eventos
n = 7

# Crear variables p1, p2, ..., pn y el multiplicador lambda
p = sp.symbols(f'p1:{n+1}')  # genera p1, p2, ..., pn
lmbda = sp.Symbol('lambda')

# Definir función entropía (log base e)
entropy_expr = -sum([p[i] * sp.log(p[i]) for i in range(n)])

# Restricción: suma de probabilidades
constraint = sum(p) - 1

# Lagrangiano
L = entropy_expr + lmbda * constraint

# Derivadas parciales
eqs = [sp.diff(L, var) for var in p] + [sp.diff(L, lmbda)]

# Resolver el sistema de ecuaciones
solution = sp.solve(eqs, p + (lmbda,), dict=True)[0]

# Mostrar solución
print("Solución óptima:")
for key in p:
    print(f"{key} = {solution[key]}")

# Entropía máxima
H_max = entropy_expr.subs(solution)
H_max_base2 = (H_max / sp.log(2)).evalf()

print(f"\nEntropía máxima (base 2): {H_max_base2}")
```
